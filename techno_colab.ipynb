```python
df_set1=df_set2
```

```python
df_set1["LoanStatus"] = df_set1["ClosedDate"].apply(lambda x: 1 if pd.isnull(x) else 0)
```

```python
df_set1["LoanStatus"] = df_set1["LoanCurrentDaysDelinquent"].apply(lambda x: 1 if x>180 else 0)
```

```python
df_set1['LoanStatus'].value_counts(dropna=False)
```

```python
from sklearn.preprocessing import LabelEncoder
def transform(feature):
    le=LabelEncoder()
    df_set1[feature]=le.fit_transform(df_set1[feature])
    print(le.classes_)
```

```python
catagorical_df=df_set1.select_dtypes(include='object')
catagorical_df.columns
```

```python
for col in catagorical_df.columns:
  transform(col)
```
